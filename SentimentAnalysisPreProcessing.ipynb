```python
import numpy as np
import pandas as pd
import re
import neattext as nt
import neattext.functions as nfx
import nltk
```

```python
data = pd.read_csv('cleaned_reviews.csv')
```

```python
data['Review'].apply(lambda x : nt.TextFrame(x).noise_scan())
```

```python
data['lowercase_normalized']=data['Review'].apply(nfx.normalize)
```

```python
data.head(10)
```

```python
data['clean_reviews']=data['lowercase_normalized'].apply(nfx.remove_emojis)
```

```python
data['clean_reviews']=data['clean_reviews'].apply(nfx.remove_special_characters)
```

```python
data['clean_reviews']=data['clean_reviews'].apply(nfx.remove_multiple_spaces)
```

```python
data['clean_reviews']=data['clean_reviews'].apply(nfx.remove_emojis)
```

```python
data.head(20)
```

```python
english_words_pattern = re.compile(r'\b\w+\b')

def remove_non_english(text):
    english_words = set(english_words_pattern.findall(text.lower()))
    return ' '.join(word for word in text.split() if word.lower() in english_words)

data['clean_reviews']=data['clean_reviews'].apply(remove_non_english)
```

```python
from langdetect import detect
```

```python
# Define a function to check if a string is English
def is_english(text):
    try:
        lang = detect(text)
        return lang == 'en'
    except:
        # If language detection fails, assume the text is non-English
        return False
```

```python
english_mask = data['clean_reviews'].apply(lambda x: is_english(str(x)))
```

```python
english_df = data[english_mask]
```
