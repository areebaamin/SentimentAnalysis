```python
import requests
from bs4 import BeautifulSoup 
```

```python
url = "https://www.tripadvisor.com/Attraction_Review-g295413-d548236-Reviews-Wagah_Border-Lahore_Punjab_Province.html"
r = requests.get(url)		# r variable has all the HTML code
```

```python

soup = BeautifulSoup(r.content, 'html.parser')

print(soup.prettify())	# to print html in tree structure
```

```python
links = []
for review in soup.findAll('a',{'class':'review_count'}):
    a = review['href']
    a = 'https://www.tripadvisor.com'+ a
    a = a[:(a.find('Reviews')+7)] + '-or{}' + a[(a.find('Reviews')+7):]
    print(a)
    links.append(a)
```
